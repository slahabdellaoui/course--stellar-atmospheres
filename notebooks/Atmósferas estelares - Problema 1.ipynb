#### B. Longitud de onda equivalente

Para una curva de sensiblidad $s\left(\lambda \right)$, la longitud de onda equivalente $\lambda_{eq}$ viene dada por la fórmula

$$
\lambda_{eq} = \frac{\displaystyle \int_{0}^{\infty} {\lambda \cdot s(\lambda) \mathrm{d} \lambda}}{\displaystyle \int_{0}^{\infty} {\lambda s(\lambda) \mathrm{d} \lambda}}.
$$

```python
from IPython.core.display import display
from IPython.core.display import HTML

def bandwidth(wavelength, intesity):
    return sum([0.5 * (i1 + i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def integrated_flux(wavelength, intesity):
    return sum([0.5 * (l1*i1 + l2*i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def equivalent_wavelength(wavelength, intesity):
    return integrated_flux(wavelength, intesity)/bandwidth(wavelength, intesity)

ordered_keys = ["U", "B", "V", "O_d", "O_n"]

headers = {
    "l": r"",
    "U": r"$U_\lambda$",
    "B": r"$B_\lambda$",
    "V": r"$V_\nu$",
    "O_d": r"$O_{dia}$",
    "O_n": r"$O_{noche}$",
}

l_eq = {
    key: equivalent_wavelength(dataset["l"], dataset[key]) for key in ordered_keys 
}

html_template = r"<table><thead><tr><td></td>{head}</tr></thead><tbody>{body}</tbody></table>"
html = HTML(
    html_template.format(
        head=r"".join([r'<td style="width: 60px;">{0}</td>'.format(headers[h]) for h in ordered_keys]),
        body=r'<tr><td  style="width: 60px;">$\lambda\left(\mu\right)$</td>{0}</tr>'.format("".join([r"<td>{0:.3f}</td>".format(v) for v in [l_eq[key] for key in ordered_keys]]))
    )
)

display(html)
```

```python
from scipy.constants import Planck
from scipy.constants import speed_of_light
from scipy.constants import Boltzmann

h = Planck
c = speed_of_light
k = Boltzmann

def planck_distribution(T, l):
    return 2 * h * (c ** 2) / ((l ** 5) * (exp(h * c / (l * k * T)) - 1))

B = planck_distribution

def B_bandwidth(wavelength, intesity, T):
    return sum([0.5 * (B(T, l1) * i1 + B(T, l2) * i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def B_integrated_flux(wavelength, intesity, T):
    return sum([0.5 * (B(T, l1) * l1 * i1 + B(T, l2) * l2 * i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def effective_wavelength(wavelength, intesity, T):
    wavelength = [x*(10**-6) for x in wavelength]
    return B_integrated_flux(wavelength, intesity, T)/B_bandwidth(wavelength, intesity, T) * (10 ** 6)

T = [25000, 10000, 5000]
l_eff = [effective_wavelength(dataset["l"], dataset["V"], t) for t in T]

ordered_keys = ["T", "l"]

headers = {
    "T": r"$T(K)$",
    "l": r"$\lambda_{eff}(\mu)$",
}

html_template = r"<table><thead><tr>{head}</tr></thead><tbody>{body}</tbody></table>"
html = HTML(
    html_template.format(
        head=r"".join([r'<td style="width: 90px;">{0}</td>'.format(headers[h]) for h in ordered_keys]),
        body=r"".join([r"<tr>{0}</tr>".format("".join([r"<td>{0:.3f}</td>".format(c) for c in r])) for r in zip(T, l_eff)])
    )
)

display(html)
```

```python
from scipy.constants import Planck
from scipy.constants import speed_of_light
from scipy.constants import Boltzmann

h = Planck
c = speed_of_light
k = Boltzmann

def planck_distribution(T, l):
    return 2 * h * (c ** 2) / ((l ** 5) * (exp(h * c / (l * k * T)) - 1))

B = planck_distribution

def bandwidth(wavelength, intesity):
    return sum([0.5 * ( i1 + i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def B_flux(wavelength, intesity, T):
    return sum([0.5 * (B(T, l1) * i1 + B(T, l2) * i2) * (l2 - l1) for l1, l2, i1, i2 in zip(wavelength[0:-1], wavelength[1:], intesity[0:-1], intesity[1:])]) 

def monocromatic_effective_flux(wavelength, intesity, T):
    wavelength = [x*(10**-6) for x in wavelength]
    return B_integrated_flux(wavelength, intesity, T)/bandwidth(wavelength, intesity)/10**12

T = [25000, 10000, 5000]
l_eff = [monocromatic_effective_flux(dataset["l"], dataset["V"], t) for t in T]

ordered_keys = ["T", "l"]

headers = {
    "T": r"$T(K)$",
    "l": r"$\left\langle B\right\rangle(kW \cdot sr^-1 \cdot m^-1 \cdot nm^-1 )$",
}

html_template = r"<table><thead><tr>{head}</tr></thead><tbody>{body}</tbody></table>"
html = HTML(
    html_template.format(
        head=r"".join([r'<td style="width: 90px;">{0}</td>'.format(headers[h]) for h in ordered_keys]),
        body=r"".join([r"<tr>{0}</tr>".format("<td>{0:.2f}</td><td>{1:.25f}</td>".format(r[0], r[1])) for r in zip(T, l_eff)])
    )
)

display(html)
```
